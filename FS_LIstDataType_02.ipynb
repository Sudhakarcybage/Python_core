```python
# organizing the list data type
cars = ["kia", "audi", "maruthi", "hyundai", "benz", "suzuki", "bmw"]
print(cars)
# temporary approach
print(sorted(cars))
print(cars)
# permanent approach
cars.sort()
print(cars)
# to find how many cars there in the list
print(len(cars))
#opposite order
cars.reverse()
print(cars)
#introduction to slicing
# syntax, list[start_value:stop_value:step_count]
students = ["divya", "harika", "naveen", "hafsa", "karthik", "vicky"]
print(students)
# last value is always exclusive, to include the last value increment the index by +1
# get divya and harika
print(students[0:2]) #['divya', 'harika']
print(students[2:4]) #['naveen', 'hafsa']
# every second person i need to get
print(students[0:6:2]) #every second value as step count is 2
```
