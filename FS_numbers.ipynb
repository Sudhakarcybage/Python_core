```python
# List variable, append(), insert(), list[index], remove(), pop(), del
num = 1
print(type(num))
x = 0.9
print(type(x))
# list
student = ["vicky",  "adhul", "shree", "anwar", "murli"]
#indexing: internally indexing will be mapped from 0, 1
# accessing the values from the list
print(student[1].title())
# how to add new element to list or modify or delete the element from list
student.append("varun")
print(student)
student.append("praneeth")
print(student)
student.insert(2, "raju")
print(student)
student[2] = "Raju" #modify
print(student)
student.remove("varun")
print(student)
#or
del student[3]
print(student) #['vicky', 'adhul', 'Raju', 'anwar', 'murli', 'praneeth']
#The .pop() method removes the last item by default but allows you to catch it and store it in a variable (like a temporary storage)
# del will remove the element from the list permanently
x = student.pop()
print(x)  #praneeth
a = student.pop(0)
print(student) #['adhul', 'Raju', 'anwar', 'murli']
#the diff btw the pop and del in the list variable
```
